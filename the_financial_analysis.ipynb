```
#@title внесение данных организации
import pandas as pd
from io import StringIO

# Считываем данные в датафрейм
TESTDATA = StringIO(
"""Object,current_period,prev_period,pre_prev_period
1150,137,61,22
1210,126,45,33
1230,13,10,2
1250,24,69,4
1310,130,130,100
1370,15,10,3
1410,72,30,10
1520,83,20,7""")

data_df = pd.read_csv(TESTDATA, index_col = 'Object')
data_df.index.name = None
```

```
#@title внесение позиций в Баланс
for cur_index in data_df.index:
  column_number=0
  for val in data_df.loc[cur_index]:
    element_index = cur_index
    element_column_name = data_df.columns.values[column_number]
    element_value = val
    # для получения следующей колонки в следующей итерации цикла
    column_number+=1
    # заполение баланса значениями
    bdf.loc[element_index, element_column_name] = element_value
```

```
#@title калькуляция позиций ИТОГО
columns = ['current_period','prev_period','pre_prev_period']
for column in columns:
  # zero to totals
  for a in range(1100,1600,100):
    bdf.loc[a,column] = 0
  for a in range(2100,2500,100):
    bdf.loc[a,column] = 0
   
  # active
  for a in range(10,100,10):
    number = 1100+a
    bdf.loc[1100,column]=bdf.loc[1100,column]+bdf.loc[number,column]
  for a in range(10,70,10):
    number = 1200+a
    bdf.loc[1200,column]=bdf.loc[1200,column]+bdf.loc[number,column]
  # total active
  bdf.loc[1600,column]=bdf.loc[1200,column]+bdf.loc[1100,column]
  # passive
  for a in range(10,80,10):
    number = 1300+a
    if a == 30: 
      continue # пропускаем, тк в балансе нет позиции 1330
    bdf.loc[1300,column]=bdf.loc[1300,column]+bdf.loc[number,column]
  for a in range(10,60,10):
    number = 1400+a
    if a == 40: 
      continue # пропускаем, тк в балансе нет позиции 1440
    bdf.loc[1400,column]=bdf.loc[1400,column]+bdf.loc[number,column]
  for a in range(10,60,10):
    number = 1500+a
    bdf.loc[1500,column]=bdf.loc[1500,column]+bdf.loc[number,column]
  #total passive
  bdf.loc[1700,column]=bdf.loc[1300,column]+bdf.loc[1400,column]+bdf.loc[1500,column]
  # валовая прибыль = выручка - себестоимость
  bdf.loc[2100,column]=bdf.loc[2110,column]-bdf.loc[2120,column]
  # прибыль от продаж = валовая прибыль - коммерческие - управленческие расходы
  bdf.loc[2200,column]=bdf.loc[2100,column]-bdf.loc[2210,column]-bdf.loc[2220,column]
  # прибыль до налогообложения
  bdf.loc[2300,column]=bdf.loc[2310,column]+bdf.loc[2320,column]-bdf.loc[2330,column]+bdf.loc[2340,column]-bdf.loc[2350,column]
  # чистая прибыль
  bdf.loc[2400,column]=bdf.loc[2300,column]-bdf.loc[2410,column]+bdf.loc[2421,column]+bdf.loc[2430,column]+bdf.loc[2450,column]+bdf.loc[2460,column]
  # совокупный финрезультат
  bdf.loc[2500,column]=bdf.loc[2400,column]+bdf.loc[2510,column]+bdf.loc[2520,column]
```
